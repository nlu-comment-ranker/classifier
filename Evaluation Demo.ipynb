### Compute the NDCG@k scores for k = 1,...,20

```python
data = data[data.set == 'test']
evaldf = evaluation.evaluate_submissions(data, 20)
evaldf.head()
```

```python
figure(figsize=(8,5))

x = range(1,21)

# All submissions
y, yerr = evaluation.calc_y_yerr(evaldf, 20)
bar(x, y, yerr=yerr, color='DarkSlateGray', ecolor='k', alpha=0.5, label="All Submissions")

# Filter to submissions with >= 3 comments
y, yerr = evaluation.calc_y_yerr(evaldf[evaldf.ncomments >= 3], 20)
bar(x, y, yerr=yerr, color='DarkSlateBlue', ecolor='k', alpha=0.5, label="3+ comments")

# Filter to submissions with >= 5 comments
y, yerr = evaluation.calc_y_yerr(evaldf[evaldf.ncomments >= 5], 20)
bar(x, y, yerr=yerr, color='ForestGreen', ecolor='k', alpha=0.5, label="5+ comments")

xlim(1,21)
ylim(0,1)

ylabel("NDCG@k")
xlabel("Number of comments considered (k)")
legend(loc='upper left', frameon=False, ncol=3)
```
